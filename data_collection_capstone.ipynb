## Download and Save Statcast Data by Year
We will retrieve Statcast pitch-level data for each MLB season (2021–2024) separately and save each as a Parquet file. This approach is memory-efficient and allows for easy reloading and merging later.

```python
data_2021 = statcast('2021-04-01', '2021-10-03')
data_2021.to_parquet('data/statcast_2021.parquet')
data_2021 = None  
```

```python
# Download 2022 Statcast data for all MLB games and save as Parquet
data_2022 = statcast('2022-04-07', '2022-10-05')
data_2022.to_parquet('data/statcast_2022.parquet')
data_2022 = None  # Free memory
```

```python
# Download 2023 Statcast data for all MLB games and save as Parquet
data_2023 = statcast('2023-03-30', '2023-10-01')
data_2023.to_parquet('data/statcast_2023.parquet')
data_2023 = None  # Free memory
```

```python
# Download 2024 Statcast data for all MLB games and save as Parquet
data_2024 = statcast('2024-03-28', '2024-09-29')
data_2024.to_parquet('data/statcast_2024.parquet')
data_2024 = None  # Free memory
```

```python
# Preview a few rows of the 2021 Statcast data to inspect columns and data structure
import pandas as pd

data_2021_preview = pd.read_parquet('data/statcast_2021.parquet')
data_2021_preview.head()
```

```python
# List of parquet files and their corresponding years
parquet_files = {
    '2021': 'data/statcast_2021.parquet',
    '2022': 'data/statcast_2022.parquet',
    '2023': 'data/statcast_2023.parquet',
    '2024': 'data/statcast_2024.parquet',
}

# Collect columns for each year
columns_by_year = {}
for year, file in parquet_files.items():
    try:
        df = pd.read_parquet(file, engine='pyarrow')
        columns_by_year[year] = set(df.columns)
    except Exception as e:
        columns_by_year[year] = set()
        print(f"Could not read {file}: {e}")

# Get all unique columns
all_columns = sorted(set.union(*columns_by_year.values()))

# Build presence matrix
presence = {
    col: [1 if col in columns_by_year[year] else 0 for year in parquet_files.keys()]
    for col in all_columns
}

# Create DataFrame for display
presence_df = pd.DataFrame.from_dict(
    presence, orient='index', columns=list(parquet_files.keys())
)
presence_df.index.name = 'column'
presence_df.reset_index(inplace=True)
presence_df
```

```python
# Filter presence_df to show only columns (rows) where at least one year is missing (contains a zero)
missing_columns_df = presence_df[(presence_df[['2021', '2022', '2023', '2024']] == 0).any(axis=1)]
missing_columns_df
```
